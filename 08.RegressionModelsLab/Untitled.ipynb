# искаме модела да е на минимално разстояние от данните
# какво е марка за разстояние
# ако от точките които сме получили (това ни е model) извадим y ще получим много на брой разлики.
# ако вземем модул на разликите, ще получим положителни числа само.
# ако осредним тази числа ще лолучим само едно число, което ние разстоянието (Loss)

```python
def regression_model(regressionData, a, b):
    plt.scatter(regressionData.x, regressionData.y)
    plt.xlabel("x: independant variable")
    plt.ylabel("y: dependant variable")


    toral_datapoints = len(regressionData.x) #очакваме x, y да имат еднаква дължина
    
    model = a * x + b
    differences = y - model
    print(model - y)
    print("----------")
    print(np.abs(model - y))
    print("----------")
    print("Loss:", np.mean(np.abs(model - y)))

    gradient_a = -2 / toral_datapoints * np.sum(x * differences) #np.sum(x * (y - model))
    gradient_b = -2 / toral_datapoints * np.sum(differences)     #np.sum(y - model)
    print("----------")
    print(gradient_a, gradient_b) # as a list
    print("----------")
    print(np.array([[gradient_a], [gradient_b]])) # as a vector
    print("----------")
    
    plt.plot(regressionData.x, model, c = "r", label = f"{a:.2f}x + {b:.2f}")
    plt.legend()
    plt.show()
```
